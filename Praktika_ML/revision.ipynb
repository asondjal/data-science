```python
import random
random.seed()

elements = [random.randint(0, 20) for _ in range(10)]

def check_element_in_list(variable: int, content: list) -> bool:
    return variable in content

def find_element_in_list(variable: int, content: list) -> str:
    print(content)
    for info in content:
        if info == variable:
            return f"Variable {variable} was found at index {content.index(info)}"
    return f"Element not found"

print(check_element_in_list(5, elements))
print(find_element_in_list(5, elements))
```

```python
# Implementierung mit zip
numbers = list(range(10))
meanings = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
values = dict(zip(numbers, meanings))
print(values)
print("Hello World!".split())
print("Hello".lower().islower())
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.1, 20, 0.1)
y = np.log(x)
plt.title("Logarithmic Funtion")
plt.xlabel("[X]: Values")
plt.ylabel("[Y]: Logarithmic Values")
plt.plot(x,y, color="orange")
plt.show()
```
